```python
import numpy as np
import pandas as pd
```

```python
month=202203
```

```python
部门费用 = pd.read_excel('费用导入模板-.xlsx', sheet_name='其它', skiprows=15, nrows= 23)
部门费用 = 部门费用.dropna(how='all').dropna(how='all', axis=1)
部门费用.columns = ['main', 'department', 'fee']
部门费用['month'] = month;
部门费用
```

```python
云服务费 = pd.read_excel('费用导入模板-.xlsx', sheet_name='其它', skiprows=2, nrows= 3)
云服务费 = 云服务费.dropna(how='all').dropna(how='all', axis=1)
云服务费.columns = ['fee_item', 'fee']
云服务费['month'] = month;
云服务费['type'] = 1;
云服务费
```

```python
技术服务费 = pd.read_excel('费用导入模板-.xlsx', sheet_name='其它', skiprows=8, nrows= 4)
技术服务费 = 技术服务费.dropna(how='all').dropna(how='all', axis=1)
技术服务费.columns = ['fee_item', 'fee']
技术服务费['month'] = month;
技术服务费['type'] = 2;
技术服务费
```

```python
运营费用 = pd.read_excel('费用导入模板-.xlsx', sheet_name='其它', skiprows=42, nrows=10)
运营费用 = 运营费用.dropna(how='all').dropna(how='all', axis=1)
运营费用.columns = ['fee_item', 'fee']
运营费用['month'] = month;
运营费用['type'] = 3;
运营费用
```

```python
    
with open('部门费用.json', 'w', encoding='utf-8') as file:
    部门费用.to_json(file, orient='records', force_ascii=False)
    
with open('云服务费.json', 'w', encoding='utf-8') as file:
    云服务费.to_json(file, orient='records', force_ascii=False)
    
with open('技术服务费.json', 'w', encoding='utf-8') as file:
    技术服务费.to_json(file, orient='records', force_ascii=False)
    
with open('运营费用.json', 'w', encoding='utf-8') as file:
    运营费用.to_json(file, orient='records', force_ascii=False)
```
